## **Analisis inicial**
### Condiciones y restricciones
Hay 30 tomas en total y tenemos la restricción de que no se pueden superar 6 tomas por día.
Por tanto, el caso óptimo no puede ser inferior a **5 días de grabación**.

Al cobrar todos los actores lo mismo por día independientemente de número de tomas, podemos **ignorar el precio por actor**.  

El **orden de las tomas y de las sesiones es irrelevante** dentro de cada sesión, ya que no afecta al gasto final. 

### (*) ¿Cuantas posibilidades hay sin tener en cuenta las restricciones?

La **única restricción** que tenemos es el máximo de tomas por día.   

Si no tenemos en cuenta esta restricción, se pueden grabar todas las tomas en un mismo día, y en tener una sola y única sesión.     

Hay que grabar **30 tomas**. Como gracias a la magia de la post-producción, el orden de las grabaciones es indiferente, podemos definir el problema como una permutación sin repetición, por lo cual tendríamos $n! = 30! \approx 2.7e32 $ combinaciones posibles de tomas dentro de esta única sesión.

### ¿Cuantas posibilidades hay teniendo en cuenta todas las restricciones?

Se define $n$ el número total de tomas y $m$ las tomas por sesión, con $n,k \in \mathbb{N}^2 $ y $1\leq k \leq n$        

Sabiendo que el orden de las grabaciones de las tomas es indiferente, elegimos $k$ veces una de las $n$ tomas para calcular el total de sesiones posibles, teniendo una menos por elegir en cada iteración.    
    
En una sesión tenemos $k$ tomas, que se pueden combinar de $k!$ maneras, por lo cual si ignoramos el orden de las tomas, tenemos una combinación sin repetición como **total de sesiones únicas**:

$$ s_{u}(n) = \frac{n!}{k! \cdot (n-k)!}$$    

Para calcular el total de soluciones, sabemos que la grabación necesitará realizar $\frac{n}{k}$ sesiones ($k \neq 0$).    
Sabemos también que el orden de las sesiones es irrelevante, y se pueden combinar de $\frac{n}{k}!$ maneras.        
Tenemos $s_{u}(n)$ sesiones por elegir para el primer día, pero las tomas no se han de repetir entre las diferentes sesiones, por lo cual tenemos $s_{u}(n-k)$ el segundo día, y así sucesivamente. Tenemos entonces, como **total de soluciones únicas posibles**:  

<br />

$$
 S = \frac{\prod_{i=0}^{ \frac{n}{k}-1} s_{u}(n-i \cdot k)}{\frac{n}{k}!}
 = \frac{1}{\frac{n}{k}!} \cdot \prod_{i=0}^{ \frac{n}{k}-1} \frac{(n-i \cdot k)!}{((n-i \cdot k) -k)! \cdot k!}
$$
      
<br /><br />
    
$$
\frac{1}{\frac{30}{6}!} *
\frac{30!}{(30-6)! \cdot 6!} *
\frac{24!}{(24 -6)! \cdot 6!} *
\frac{18!}{(18 -6)! \cdot 6!} *
\frac{12!}{(12 -6)! \cdot 6!} *
\frac{6!}{(6 -6)! \cdot 6!}
\approx 1,14 \cdot10^{16}
$$

````
import numpy as np
import pandas as pd

def calculate_cost(solution : tuple, max_takes: int, actores_tomas : np.array) -> int:
  # Calculamos el número de sesiones que se necesitarán
  num_ses = len(solution)

  # Inicializamos costes a 0
  solution_cost = 0

  # Recorremos cada sesión
  for session in solution:
    
    takes_cnt = 0
    session_cost = 0

    # Sumamos las tomas de cada actor para ver si participa en alguna
    actores_tomas_session = actores_tomas[session]
    suma_actores = np.sum(actores_tomas_session, axis=0)

    # Si un actor ha participado en alguna toma de la sesión 
    # sumamos 1 al costo de la sesión
    for actor in range(0, len(suma_actores)):
      if suma_actores[actor] > 0:
        session_cost = session_cost + 1
    # Sumamos el costo de la sesión al total
    solution_cost = solution_cost + session_cost
      
  return solution_cost
````

### Diseña un algoritmo para resolver el problema por fuerza bruta

Se diseña el algoritmo para resolver el problema por fuerza bruta pero se lanzará el script con un volumen de datos menor dado su gran coste computacional:

````
def brute_force(actores_tomas : np.array, max_takes: int, takes) -> tuple:
    solutions = todas_las_combinaciones(max_takes, takes)

    best_solution = None
    best_solution_cost = None
    # Buscamos entre todas las soluciones para encontrar la de menor coste
    for solution in solutions:
        solution_cost = calculate_cost(solution, max_takes, actores_tomas)
        if best_solution is None or solution_cost < best_solution_cost:
            best_solution = solution
            best_solution_cost = solution_cost
    return best_solution_cost, best_solution

# Calculamos todas las posibles combinaciones de las tomas
def permutation(lst):
 
    if len(lst) == 0:
        return []
    if len(lst) == 1:
        return [lst]
    l = []
    
    for i in range(len(lst)):
       m = lst[i]
       remLst = lst[:i] + lst[i+1:]
       
       for p in permutation(remLst):
           l.append([m] + p)
    return l

def todas_las_combinaciones(max_takes, takes):
    takes_filt = []
    for p in permutation(takes):
        output=[p[i:i + max_takes] for i in range(0, len(p), max_takes)]
        for i in range(len(output)):
          output[i].sort()
        if output not in takes_filt:
          takes_filt.append(output)
    return takes_filt  
````

### Calcula la complejidad del algoritmo por fuerza bruta

Como el algoritmo por fuerza bruta genera todas las soluciones posibles,  $n!$  soluciones, tiene una complejidad  $O(n!)$

### (*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

Lo primero a tener en cuenta es que lo ideal sería que en cada sesión, los actores que vayan puedan grabar el máximo de sus tomas totales, por lo que queremos conseguir es conseguir una lista de las tomas ordenadas por sesiones (de 6 en 6 tomas) empezando por las que más actores participan:

````
from operator import itemgetter
from heapq import nlargest
import random

# Aplicamos el algoritmo voraz
def alg_voraz(actores_tomas, max_takes, total_act_toma, intentos, iteracion = 0, best_result = 0, best_solution = 0):

    # Para la primera iteración
    if (iteracion == 0):
        # Hacemos una copia de la lista de actores por toma para que no se modifique
        total_act_toma_cpy = total_act_toma.copy()

        # Ordenamos solución que agrupa las sesiones por costes para tener una primera aproximación
        solution = ordenar_tomas(actores_tomas, max_takes, total_act_toma_cpy, False)
        # Calculamos los costes de la primera ordenación
        cost = calculate_cost(solution, max_takes, actores_tomas)

        # Al ser la primera iteración, será nuestra mejor solución por el momento
        best_result =  cost
        best_solution = solution

        iteracion += 1
        print(f"iteracion {iteracion}")
        print(f"Solucion parcial = {solution}")
        print(f"Coste parcial = {cost}")
        # Volvemos a iterar
        alg_voraz(actores_tomas, max_takes, total_act_toma, intentos, iteracion, best_result, best_solution)

    # En las siguientes iteraciones
    elif (iteracion < intentos):
        total_act_toma_cpy = total_act_toma.copy()

        # Mutamos la mejor solución hasta el momento para ver si conseguimos una mejor
        solution = mutation(best_solution, len(actores_tomas)/max_takes, max_takes)
        cost = calculate_cost(solution, max_takes, actores_tomas)

        iteracion += 1
        print(f"iteracion {iteracion}")
        print(f"Solucion parcial = {solution}")
        print(f"Coste parcial = {cost}")

        # Si hemos encontrado una solución con menos coste, la guardamos
        if cost < best_result:
          best_result = cost
          best_solution = solution
        print(f"Mejor Coste = {best_result},\nMejor Solution = {best_solution}")
        
        # Iteramos de nuevo
        alg_voraz(actores_tomas, max_takes, total_act_toma, intentos, iteracion, best_result, best_solution)

    # Cuando hayamos acabado las iteraciones, acabamos el programa
    else:
        return best_result, best_solution

# Función que ordenará las tomas según su participación de actores (de más a menos)
def ordenar_tomas(actores_tomas, max_takes, total_act_toma, mutation = False, lista_ordenada_tomas = [], session = 1):
    # Habrá tantas sesiones como tomas_totales/max_tomas_por_dia
    if (session <= int(len(actores_tomas)/max_takes)):
      lista_tomas = []
      # Cogemos los valores con mayor número de actores por toma
      largest = nlargest(max_takes, enumerate(total_act_toma), itemgetter(1))
      
      # Lo recorremos para quedarnos con el índice (la toma) y ponemos un 0 en su posición para que no
      # la tengamos en cuenta en su siguiente iteración
      for i in range(len(largest)):
        lista_tomas.append(largest[i][0])
        total_act_toma[largest[i][0]] = 0
      # Aumentamos sesión
      session += 1
      # Metemos la lista ordenada en la sublista final
      lista_ordenada_tomas.append(lista_tomas)
      # Volvemos a buscar los mayores valores para sacar las tomas sin contar esta vez con las que ya hemos sacado
      ordenar_tomas(actores_tomas, max_takes, total_act_toma, mutation, lista_ordenada_tomas, session)

    # Tras la última sesión, devolvemos la lista de listas de tomas ordenadas por costes
    return lista_ordenada_tomas

# Función que mutará nuestra mejor solución para encontrar una posible mejor solución
def mutation(solution, sessions, max_takes):
    # Elegimos dos números random sin repetición para intercambiar dos tomas y dos sesiones
    rand_sess = random.sample(range(int(sessions)), 2) 
    rand_toma = random.sample(range(max_takes), 2)
    
    # Hacemos el intercambio
    tmp = solution[rand_sess[0]][rand_toma[0]]
    solution[rand_sess[0]][rand_toma[0]] = solution[rand_sess[1]][rand_toma[1]]
    solution[rand_sess[1]][rand_toma[1]] = tmp

    return solution
````

### (*)Calcula la complejidad del algoritmo 

La complejidad de la función "ordenar_tomas" es la complejidad de "nlargest": $O(m * log(t))$ siendo m = máximo tomas por día y t el número de tomas totales, repitiéndose el número de sesiones (tomas totales/máximo tomas por día) por lo que se nos queda $O(m^2 * log(t))$.

La complejidad de la función "mutation" es de $O(1)$

La complejidad de la función final, "alg_vor" es $n$ (siendo $n$ el número de iteraciones) multiplicado por las dos funciones que aplican, por lo que la complejidad se nos queda:
    
**Complejidad algoritmo** = $O(n * n^2 * log(t) * 1)$ = $O(n^3*log(t))$ = $O(n^3)$

````
import random
import numpy as np

# Máximo de tomas para el juego de datos.
total_takes = random.randint(10, 50)

# Número máximo de sesiones.
max_takes = random.randint(3, int(total_takes / 3))

# Restricción de número de tomas por sesión.
session_takes = random.randint(2, int(total_takes / max_takes))

# Número de tomas.
takes = session_takes * max_takes

# Número de actores a tener en cuenta
n_actors = random.randint(4, 20)

# Probabilidad de actuar en una toma (25%)
p = 0.25 

# Generamos las participaciones de los actores por toma
actores_tomas = ()
for take_id in range(takes):
    take = []
    for actor in range(n_actors):
        take.append(int(random.random() < p))

    # Nos aseguramos tener al menos 1 actor por toma.
    if not sum(take):
        actor = random.randint(0, n_actors - 1)
        take[actor] = 1

    actores_tomas += (take,)

total_act_toma = np.zeros(shape=(len(actores_tomas)))

for i in range(len(actores_tomas)):
    total_act_toma[i] = sum(actores_tomas[i])

actores_tomas = np.array(actores_tomas)

print('Tomas totales:', takes)
print('Tomas máximas por sesión:', max_takes)
print('Total de actores:', n_actors)
print('Total actores por toma:', total_act_toma)
print('Actores_tomas:')
display(actores_tomas)
````
